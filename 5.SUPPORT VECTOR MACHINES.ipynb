## SVR BASED REGRESSION MODEL

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C:/Users/ADMIN/Desktop/python jupyter/DATA SCIENCE/CSV FILES/Movie_regression.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(),inplace=True)
```

```python
df.info()
```

```python
df = pd.get_dummies(df,columns = ["3D_available","Genre"],drop_first=True)
```

```python
df.head()
```

```python
X = df.loc[:,df.columns!="Collection"]
Y = df['Collection']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

## Standardizing Data
### https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_train_std
```

```python
X_test_std
```

## training a regression tree
### https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear',C=5000)
```

```python
svr.fit(X_train_std,Y_train)
```

```python
Y_pred = svr.predict(X_test_std)
Y_pred
```

## model performance

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(Y_test,Y_pred)
```

```python
r2_score(Y_test,Y_pred)
```

# SVM BASED CLASSIFICATION MODEL

```python
df = pd.read_csv('C:/Users/ADMIN/Desktop/python jupyter/DATA SCIENCE/CSV FILES/Movie_classification.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(),inplace=True)
```

```python
df = pd.get_dummies(df,columns = ["3D_available","Genre"],drop_first=True)
```

```python
df.head()
```

```python
X = df.drop(['Start_Tech_Oscar'],axis=1)
Y = df['Start_Tech_Oscar']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
sc = StandardScaler().fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_train_std
```

```python
X_test_std
```

## training svm : kernel = linear
### https://scikit-learn.org/stable/modules/svm.html

```python
from sklearn import svm
```

```python
clf_svm = svm.SVC(kernel='linear',C=0.01)
clf_svm.fit(X_train_std,Y_train)
```

```python
Y_pred = clf_svm.predict(X_test_std)
Y_pred
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
clf_svm.n_support_
```

## kernel = polynomial

```python
clf_svm_poly = svm.SVC(kernel='poly',degree=2,C=0.01)
clf_svm_poly.fit(X_train_std,Y_train)
```

```python
Y_pred = clf_svm_poly.predict(X_test_std)
Y_pred
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
clf_svm_poly.n_support_
```

## kernel = radial

```python
clf_svm_rad = svm.SVC(kernel='rbf',gamma=0.5,C=10)
clf_svm_rad.fit(X_train_std,Y_train)
```

```python
Y_pred = clf_svm_rad.predict(X_test_std)
Y_pred
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
clf_svm_rad.n_support_
```

# HYPERPARAMETER TUNING
## Linear GRID SEARCH CV

```python
from sklearn.model_selection import GridSearchCV 
```

```python
params = {'C':(0.001,0.005,0.01,0.05,0.1,0.5,1,5,10,50,100,500,1000)}
```

```python
clf_svm1 = svm.SVC(kernel='linear')
```

```python
svm_grid_lin = GridSearchCV(clf_svm1,params,n_jobs=-1,cv=10,verbose=1,scoring='accuracy')
```

```python
svm_grid_lin.fit(X_train_std,Y_train)
```

```python
svm_grid_lin.best_params_
```

```python
lin_svm_clf = svm_grid_lin.best_estimator_
```

```python
Y_pred = lin_svm_clf.predict(X_test_std)
Y_pred
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```
