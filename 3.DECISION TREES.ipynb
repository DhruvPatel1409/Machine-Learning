# REGRESSION TREE

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C:/Users/ADMIN/Desktop/python jupyter/DATA SCIENCE/CSV FILES/Movie_regression.csv')
```

```python
df.head()
```

```python
df.info()
```

# missing value imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(),inplace=True)
```

```python
df.info()
```

# dummy variable creation

```python
df.head()
```

```python
df = pd.get_dummies(df,columns = ["3D_available","Genre"],drop_first=True)
```

```python
df.head()
```

```python
X = df.loc[:,df.columns!="Collection"]
Y = df['Collection']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

# training a regression tree

```python
from sklearn import tree
```

```python
regtree = tree.DecisionTreeRegressor(max_depth=3)
```

```python
regtree.fit(X_train,Y_train)
```

```python
Y_pred = regtree.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print(mean_squared_error(Y_test,Y_pred))
print(r2_score(Y_test,Y_pred))
```

# plotting a decision tree

```python
dot_data = tree.export_graphviz(regtree,out_file=None)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

# pruning a decison tree
# There are 3 ways for controlling tree growth :
# 1) maximum number of levels in tree

```python
regtree1 = tree.DecisionTreeRegressor(max_depth=3)
regtree1.fit(X_train,Y_train)
dot_data = tree.export_graphviz(regtree1,out_file=None,feature_names=X_train.columns,filled=True)
graph1 = pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())
```

# 2) minimum observations at internal node

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split=40)
regtree2.fit(X_train,Y_train)
dot_data = tree.export_graphviz(regtree2,out_file=None,feature_names=X_train.columns,filled=True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

# 3) minimum obsrevations at leaf node

```python
regtree3 = tree.DecisionTreeRegressor(min_samples_leaf=35)
regtree3.fit(X_train,Y_train)
dot_data = tree.export_graphviz(regtree3,out_file=None,feature_names=X_train.columns,filled=True)
graph3 = pydotplus.graph_from_dot_data(dot_data)
Image(graph3.create_png())
```

# CLASSIFICATION TREE

```python
df = pd.read_csv('C:/Users/ADMIN/Desktop/python jupyter/DATA SCIENCE/CSV FILES/Movie_classification.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(),inplace=True)
```

```python
df = pd.get_dummies(df,columns = ["3D_available","Genre"],drop_first=True)
```

```python
df.head()
```

```python
X = df.drop(['Start_Tech_Oscar'],axis=1)
Y = df['Start_Tech_Oscar']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

# training classification tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
clftree = DecisionTreeClassifier(max_depth=4)
```

```python
clftree.fit(X_train,Y_train)
```

```python
Y_pred = clftree.predict(X_test)
Y_pred
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```

# plotting a classification tree

```python
dot_data = tree.export_graphviz(clftree,out_file=None,feature_names=X_train.columns,filled=True)
```

```python
from IPython.display import Image
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

# controlling tree growth

```python
clftree1 = DecisionTreeClassifier(max_depth=4,min_samples_leaf=20)
clftree1.fit(X_train,Y_train)
dot_data = tree.export_graphviz(clftree1,out_file=None,feature_names=X_train.columns,filled=True)
graph1 = pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())
```

```python
Y_pred = clftree1.predict(X_test)
Y_pred
```

```python
accuracy_score(Y_test,Y_pred)
```
