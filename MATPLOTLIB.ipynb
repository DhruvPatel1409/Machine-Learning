```python
fig,axes = plt.subplots(dpi=150)
axes.plot(x,x+1,color='red',alpha=0.5,linewidth=0.25)
axes.plot(x,x+2,color='green',alpha=0.75,linewidth=0.55)
axes.plot(x,x+3,color='black',alpha=0.95,linewidth=1.75)
axes.plot(x,x+4,color='orange',alpha=0.95,linewidth=2)
```

```python
fig, axes = plt.subplots(dpi=150)

axes.plot(x, x+1, color='blue', linewidth=0.25, linestyle='-')
axes.plot(x, x+2, color='blue', linewidth=0.55, linestyle='-.')
axes.plot(x, x+3, color='blue', linewidth=1.75, linestyle=':')

line, = axes.plot(x, x+4, color='black', linewidth=1.50)
line.set_dashes([5, 10, 15, 10])

line, = axes.plot(x, x+5, color='black', linewidth=2)
line.set_dashes([5, 10, 15, 3])

line, = axes.plot(x, x+6, color='black', linewidth=2)
line.set_dashes([5, 10, 4, 10])

plt.show()
```

```python
fig, axes = plt.subplots(dpi=100)

axes.plot(x, x+1, color='blue',marker='o')
axes.plot(x, x+2, color='blue', marker='+')
axes.plot(x, x+3, color='blue',marker='s')
axes.plot(x, x+4, color='blue',marker='^')

axes.plot(x,x+5,color='black',marker='o',markersize=5)
axes.plot(x,x+6,color='black',marker='o',markerfacecolor='red')
axes.plot(x,x+7,color='black',marker='o',markersize=15,markerfacecolor='yellow')

plt.show()
```

# other 2D graphics

```python
n = np.array([0,1,2,3,4,5])

fig,axes = plt.subplots(1,4,figsize=(16,5))
axes[0].set_title('scatter plot')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n,n**2,lw=2)

axes[2].set_title('bar')
axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.5)

axes[3].set_title('fill between')
axes[3].fill_between(x,x**2,x**3,color='green',alpha=0.5)
```

# radial plot

# histogram

```python
n = np.random.randn(100000)
fig,axes = plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("histogram")
axes[0].hist(n)

axes[1].set_title("cumulative detailed histogram")
axes[1].hist(n,cumulative=True,bins=50)
```

# contour image

```python
delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2-Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2                           
```

```python
print(X)
print(Y)
```

```python
fig, axes = plt.subplots()
CS = axes.contour(X, Y, Z)
axes.clabel(CS, inline=1, fontsize=10)
axes.set_title('Contour Map')
plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(X,Y,Z,rstride=4,cstride=4,lw=0)
```

# pie graph

```python
lang = ['Python', 'Java', 'C++', 'C', 'Js', 'C#']
pop = [22.2, 17.6, 8.8, 8, 7.7, 6]
colors = ['red', 'green', 'blue', 'yellow', 'cyan', 'whitesmoke']
explode = (0.1, 0, 0, 0, 0, 0)

plt.pie(pop, explode=explode, labels=lang, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Popularity of Programming Languages')
plt.show()
```
